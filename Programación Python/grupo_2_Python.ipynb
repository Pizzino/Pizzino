7. Obtener el estimador de la varianza del termino de perturbación

```python
import random 

random.seed(175)

x1 = np.random.rand(500) # uniform distribution  [0,1]
x2 = np.random.rand(500) # uniform distribution [0,1]
x3 = np.random.rand(500) # uniform distribution [0,1]
x4 = np.random.rand(500) # uniform distribution [0,1]
e = np.random.normal(0,1,500) # normal distribution mean = 0 and sd = 1

# Poblacional regression (Data Generating Process GDP)


Y = 1 + 0.8*x1 + 1.2*x2 + 0.5*x3 + 1.5*x4 + e
```

```python
X = np.column_stack((np.ones(500),x1,x2,x3,x4))
X
```

```python
beta = np.linalg.inv(X.T @ X) @ ((X.T) @ Y )
beta
```

```python
Y_hat = X@beta
Y_hat
```

```python
U_hat = Y - Y_hat
U_hat
```

```python
U_hat.T
```

```python
U_hat.T@U_hat
```

```python
SCR = U_hat.T@U_hat
SCR
```

```python
Estimador_de_la_varianza_del_termino_de_pertubacion = SCR/(500-5)
Estimador_de_la_varianza_del_termino_de_pertubacion
```

8. Matriz de varianza y covarianza de los coeficientes

```python
Estimador_de_la_varianza_del_termino_de_pertubacion*np.linalg.inv(X.T @ X)
```
