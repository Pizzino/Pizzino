9. Crear una base de datos con la siguiente información:

a. Quedarse con la información de los distritos de Ciudad del Cerro, Jauja, Acolla, San Gerónimo, Tarma, Oroya y Concepción

b. Ciudades con una cantidad de niños (4-6 años) mayor a 50 y de (6-14 años) > 70

c. Ordenar las variables según el total de la población y cantidad de población nativa

d. Solo quedarse con las variables trabajadas en 8, Distrito, y comunidad.

e. Guardar la mitad de la información en bases diferentes (llamar a cada uno Base1 y Base 2)

f. Guardar la Base1 en formato csv y excel en la carpeta data. (Use el siguiente nombre Base1_WG(numero de grupo))

```python
# a. Nos quedamos con la información solicitada
base=junin_data[(junin_data["District"] == "CIUDAD DEL CERRO")|(junin_data["District"] == "CIUDAD DEL CERRO")|
           (junin_data["District"] == "JAUJA")|(junin_data["District"] == "ACOLLA")|
           (junin_data["District"] == "SAN GERÓNIMO")|(junin_data["District"] == "TARMA")|
           (junin_data["District"] == "OROYA")|(junin_data["District"] == "CONCEPCIÓN")]       

# b. Nos quedamos con las ciudades con cantidades mayores a las solicitadas
Base=base[(base["4_6_years_total"]> 50 ) & (base["6_14_years_total"]> 70 )]       
Base
```

10. Crear una función que me permita extraer un rango de columnas y de filas. Crear una función que me permite extraer una columna según el nombre de la variable y dos filas.

```python
#Funcion que selecciona rango de columnas y filas
def select(a,cname1,cname2,f1,f2): #select(Base de datos, nombre de columna1, nombre columna2, rango 1, rango2)
    y = a.loc[f1:f2,(cname1,cname2)]
    return y
```

```python
select(junin_data,'comunidad','natives',1,5)
```

```python
#Funcion que extrae columnas por el nombre
def extraer(a,cname,f1,f2): #select(Base de datos, nombre de columna1, rango 1, rango2)
    y = a.loc[f1:f2,cname]
    return y
```
