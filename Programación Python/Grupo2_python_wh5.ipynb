```python
import pandas as pd
import numpy as np
import chardet # to get string character format 
import re
```

```python
#Data ENAHO PANEL 2015-2019
data1 = pd.read_stata("C:/Users/USER/Documents/GitHub/1ECO35_R_PYTHON/data/Juntos_program/data.dta")
#Data Programa Juntos
data2 = pd.read_stata("C:/Users/USER/Documents/GitHub/1ECO35_R_PYTHON/data/Juntos_program/unidos.dta")
```

### Replicación Clean DHS

```python
data1.columns
```

```python
filter_list = []
filter_list1 = []

#Drop special character '_'
for i in range(0,len(list(data1.columns))):

        w =  re.sub('[\_]','', list(data1.columns)[i])
        filter_list.append(w)
        
        
#Drop last two string
for i in range(0,len(list(filter_list))):

        w =  re.sub(".{2}$",'', list(filter_list)[i])
        filter_list1.append(w)
```

```python
# drop duplicates in a list

new_list = list(dict.fromkeys(filter_list1))
new_list
```

#### Data ENAHO Panel

```python
reshape_panel = pd.wide_to_long(data1, stubnames = new_list, i = ['numper'] , j = 'year' , sep = '_').reset_index()
```

```python
reshape_panel
```

```python
new_list[0] = data1.columns[0]
new_list[1] = data1.columns[1]
new_list.append("year")
data_panel = reshape_panel[new_list]
data_panel = data_panel.rename(columns={'numpanh15': 'numpanh'})
data_panel                               
```

#### MERGE

```python
data2
```

```python
#### Combinamos las bases de PANEL ENAHO y programa Juntos
base_total = pd.merge(data_panel,data2,on='ubigeo')
```

```python
base_total
```

### LABELS

```python
# creando labels mediante diccionario:

labels = { 
    'numper': 'Unique individual identifier', 
    'numpanh' : 'Unique household identifier in 2015', 
    'mes' : 'Month of the survey in 2015', 
    'p400a3_15' : 'Year of birth of the person, reported in 2015', 
    'p4022' : { '1' : 'if the person was sick in the last 4 weeks', '0' : 'if the person was not sick in the last 4 weeks'},
    'unidos' : {'1' : 'Porgrama Juntos applied in the district', '0' : 'Programa Juntos was not applied in the district'}
}
```

### PORCENTAJE

```python
#Obtener una base de datos a nivel departamentos con el porcentaje  de dsitritos beneficiados por el programa juntos.

Base_1 = pd.DataFrame(base_total.groupby( ["year","region"] , as_index = False).agg(total_beneficiados = ('unidos', np.size),total_beneficiados_distrito = ('unidos', np.sum)))
Base_1['Porcentaje_beneficiados_distrito']=Base_1.total_beneficiados_distrito/Base_1.total_beneficiados
Base_1
```
