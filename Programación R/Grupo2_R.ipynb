Simule un proceso generador de datos con 10 mil observaciones como se explicó en la primera práctica dirigida. Use un loop para obtener los coeficientes estimados con los siguientes tamaños de muestra (10, 50, 100,200,500,1000,5000)


```r
set.seed(36121)

x1 <- runif(10000)
x2 <- runif(10000)
x3 <- runif(10000)
x4 <- runif(10000)
e <- rnorm(10000)

# Poblacional regression (Data Generating Process GDP)

Y <- 1 + 0.3*x1 + 0.75*x2 - 4.3*x3 + 3.6*x4 + e
```

```r
# Matriz X de covariables
X <- cbind(matrix(1,10000), x1,x2,x3,x4)
X
```

```r
beta <- solve(t(X) %*% X) %*% (t(X) %*% Y)
beta
```

```r
#Ahora aplicamos loops
num <- c(10,50,100,200,500,1000,5000)
betas <- matrix(0,5,7)
for (i in 1:length(num)){
    Xn <- X[1:num[i],]
    betaN <- solve(t(Xn) %*% Xn) %*% (t(Xn) %*% Y[1:num[i]])
    betas[,i] <- betaN
    #print(num[i])
    #print(betaN)
}
#print(betas)
#print(cbind(num,t(betas)))
matrix_betas <- as.data.frame(cbind(num,t(betas)))
names(matrix_betas) <- c("N","b1","b2","b3","b4","b5")
matrix_betas
```

Use la base de datos production_fish.csv para obtener el porcentaje de consumo de ciertos tipos de producción marina respecto al consumo de anchoveta. Crear esta información solo para los siguiente tipos producción marina: Atun, Jurel, Langostino, Tollo.

```r
library("glue")
```

```r
production_fish = read.csv("../../data/production_fish.csv" )
```

```r
head(production_fish)
```

```r
names(production_fish)
```

```r
select <- c("Atun","Jurel","Langostino","Tollo")
for (col in select){
    production_fish[glue('{col}_%')]   <- (production_fish[glue('{col}')]  * 100) / production_fish$Anchoveta
    }
head(production_fish)
```

Crear una matrix aleatoria de orden 100x10. Crear un función que permita estandariza cada columna; es decir, cada observación de una columna debe restarse por su media y dividirse por la desviación estándar de la columna. Llame a esta función como standarize

```r
set.seed(71999)
x1 <- sample(100)
```

```r
Mat <- matrix(x1,nrow=100, ncol=10, byrow = T)
Mat
```

```r
media <- mean(Mat)
media
```

```r
stand <- sd(Mat)
stand
```

```r
standarize <- function (Mat){
    z <- ( Mat- mean(Mat) ) / sd(Mat)
}
```

```r
std_Mat <- apply(Mat, 2, standarize)
std_Mat
```
