8. Crear columnas con las siguiente información: el % de mujeres que no escriben ni leen, % de varones que no escriben ni leen y % de nativos respecto al total de la población

```r
junin_data3['%nesc_mujer']=junin_data3$nesc_mujer/junin_data3$total_not_write #creamos la variable que mide el % de mujeres que no escriben
junin_data3['%nlee_mujer']=junin_data3$nlee_mujer/junin_data3$total_not_read #creamos la variable que mide el % de mujeres que no leen
junin_data3['%nesc_hombre']=junin_data3$nesc_hombre/junin_data3$total_not_write #creamos la variable que mide el % de hombres que no escriben
junin_data3['%nlee_hombre']=junin_data3$nlee_hombre/junin_data3$total_not_read #creamos la variable que mide el % de hombres que no leen
junin_data3['%nativos']=junin_data3$natives/(junin_data3$whites+junin_data3$natives+junin_data3$mestizos+junin_data3$blacks) #creamos la variable que mide el % de nativos
junin_data3['poblacion']=(junin_data3$whites+junin_data3$natives+junin_data3$mestizos+junin_data3$blacks) #creamos la variable que mide la población total
junin_data3
```

9. Crear una base de datos con la siguiente información:

a. Quedarse con la información de los distritos de Ciudad del Cerro, Jauja, Acolla, San Gerónimo, Tarma, Oroya y Concepción

b. Ciudades con una cantidad de niños (4-6 años) mayor a 50 y de (6-14 años) > 70

c. Ordenar las variables según el total de la población y cantidad de población nativa

d. Solo quedarse con las variables trabajadas en 8, Distrito, y comunidad.

e. Guardar la mitad de la información en bases diferentes (llamar a cada uno Base1 y Base 2)

f. Guardar la Base1 en formato csv y excel en la carpeta data. (Use el siguiente nombre Base1_WG(numero de grupo))

```r
# a. Nos quedamos con la información solicitada
base <- junin_data3[junin_data3$District %in% c("CIUDAD DEL CERRO",
                                                    "JAUJA",
                                                    "ACOLLA",
                                                    "SAN GERÓNIMO",
                                                    "TARMA",
                                                    "OROYA",
                                                    "CONCEPCIÓN"),]

base
```

```r
# b. Nos quedamos con las ciudades con cantidades mayores a las solicitadas
Base <- base[which(base['4_6_years_total'] > 50 &
                             base['6_14_years_total'] > 70), ]
Base
```

```r
#c. ordenamos la base de datos de acuerdo al número de poblacion y nativos por distrito
BASE <- Base[order(Base$poblacion,Base$natives),]
BASE
```

```r
#d. Nos quedamos con las variables importantes
data<-BASE[,c('District','comunidad','%nesc_mujer','%nlee_mujer','%nesc_hombre','%nlee_hombre','%nativos','poblacion')]
data
```

```r
#e. Dividimos los datos en dos partes
Base1<-data[0:5,]
Base1
```

 10. Crear una función que me permita extraer un rango de columnas y de filas. Crear una función que me permite extraer una columna según el nombre de la variable y dos filas.

```r
#Funcion que extrae rango de columnas y filas
extraer<- function(a,f1,f2,c1,c2){ #select(Base de datos, fila 1, fila2, col1,col2)
    y = a[f1:f2,c1:c2]
    return (y)
}
```

```r
extraer(junin_data,1,2,2,3)
```

```r
#Funcion que extrae columnas por el nombre
#select(Base de datos, nombre de columna, rango 1, rango2)
select <- function(a,name,b,c){
    y = a[b:c,c(name)]
    return (y)
}
```

```r
select(junin_data,"District",1,2)
```
