4.-Group by.

Obtener el total de hogares por distrito, el total de hogares que cuentan con internet por distrito, finalmente el porcentaje de hogares que cuentan con internet.

```r
data <- example_2 %>% group_by(CCDI) %>% summarize(total_hogares_distrito = n(), total_hogares_internet=sum(Dummy_2))
data
#Creamos una base de datos que agrupe por distritos y agregamos las variables solicitadas.
```

```r
data$Porcentaje_hogares_con_internet <- round(((data$total_hogares_internet)/(data$total_hogares_distrito)), digit=2)
data
```

Obtener una base de datos a nivel departamentos con la siguiente informaci√≥n: total de hogares, total de hogares cuyo vecindario instal√≥ c√°maras de seguridad y el porcentaje hogares cuyos vecindarios cuentan con c√°maras de seguridad. Base_1

```r
Base_1 <- example_2 %>% group_by(CCDD) %>% summarize(total_hogares_distrito = n(), total_hogares_camaras=sum(!is.na(Dummy_1)))
Base_1
```

```r
Base_1$Porcentaje_hogares_camaras<- (Base_1$total_hogares_camaras)/(Base_1$total_hogares_distrito)
Base_1
```

Crear una base de datos a nivel departamentos con la siguiente informaci√≥n: total de hogares que cuentan con t√≠tulos de propiedad y porcentaje de hogares que cuentan con t√≠tulos de propiedad. Base_2
